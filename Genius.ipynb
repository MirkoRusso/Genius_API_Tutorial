## 5.1 Sentiment Analisys    
![Sentiment](immagini/Sentiment.png) 
_L'analisi del sentiment è una tecnica di Natural Language Processing, che ci consente di determinare se il tono emotivo di un testo è **positivo, negativo o neutro.** Gli strumenti di analisi del sentiment possono scansionare questo testo per determinare automaticamente l’atteggiamento dell’autore nei confronti di un argomento._

Le classi determinate sono le seguenti:
1. **Fine-grained :**   
    1.1  Very positive   
    1.2 Positive  
    1.3 Neutral   
    1.4 Negative  
    1.5 Very negative   

2. **Emotion detection :** usata per capire le emozioni all’interno del testo.  
3. **Aspect based :** focalizzata sul cogliere un aspetto della frase   
4. **Multilingual :** sentiment analysis fatta su testi scritti in più lingue       

Nella Sentiment Analisys si rappresentano le parole come simboli e ne viene rimosso il significato semantico.    
Si vuole una misura di distanza semantica tra le parole di documenti diversi perchè si vuole che parole vicine abbiano lo stesso significato e distanza più grandi facciano riferimento a termini lontani semanticamente. 

Consideriamo il contesto della frase.   
Si suddivide la frase in token e data una parola si cerca di capirne il contesto. Si fissa una finestra di ampiezza opportuna fissata a priori e si centra la finestra sulla parola interessata. Il contesto è quindi l’iniseme di parole che tiriamo fuori utilizzando la finestra.       

Si cerca di capire quanto due parole siano simili ponendo l’una nel contesto dell’altra poichè si pensa che parole simili abbiano contesti simili.

Qui entrano in gioco i termini: **Matrice di Co-occorrenza** e **Word Embedding**
- **Matrice di Co-occorenza :** 
E' uno strumento utilizzato per rappresentare la frequenza con cui le coppie di termini appaiono insieme in un dato insieme di documenti. Questa matrice può aiutare a catturare le relazioni tra le parole e a comprendere meglio il contesto in cui appaiono.

- **Word Embedding :**  
Tecnica utilizzata nell'ambito del Natural Language Processing per rappresentare le parole come vettori numerici in uno spazio multidimensionale. Questa rappresentazione vettoriale consente di catturare le relazioni semantiche tra le parole, rendendo possibile l'analisi e l'elaborazione automatica del linguaggio naturale da parte dei computer.       

![WordEmbedding](immagini/WordEmbedding.png)

Una volta introdotti questi termini, possiamo parlare di **GloVe**  
Acronimo di _**Global Vectors for Word Representation**_, è un algoritmo di embedding di parole che si basa sulla co-occorrenza delle parole in un corpus di testi.     
La co-occorrenza si riferisce alla frequenza con cui due parole compaiono insieme in un contesto specifico o in un insieme di documenti.    
L'obiettivo principale di **GloVe** è quello di rappresentare le parole in modo che la similarità semantica tra di esse sia riflessa nello spazio vettoriale.       
L'intuizione chiave di **GloVe** è che le informazioni semantiche e sintattiche di una parola possono essere catturate esaminando come la parola co-occorre con altre parole.   
La frequenza di co-occorrenza di due parole riflette il grado di relazione tra di loro.  
Ad esempio, parole che sono spesso vicine l'una all'altra in contesti diversi avranno una relazione semanticamente più forte.   
**GloVe** sfrutta questa intuizione cercando di apprendere delle rappresentazioni vettoriali in modo che il loro prodotto scalare rifletta il rapporto di probabilità delle co-occorrenze delle parole. 
*Si considerano i seguenti elementi :*
- Il word embedding *Wi* per ogni parametro *i* in input del modello.
- Due parole caratterizzate da *Pij*, ovvero la probabilità che *j* appaia nel contesto di *i*
- Se il termine *i* e il termine *j* sono simili e lo sono anche *j* e *k* allora *i* e *k* sono simili per transitività tra loro.
- Il prodotto vettoriale tra *Wi* e *Wj* mi da uno scalare *Pij* che quantifica la relazione tra le parole *i* e *j*.

Dopo questa breve introduzione alla Sentiment Analisys, possiamo passare alla visione del modello **VADER** e al suo funzionamento.

## 5.2 Cos'è Vader?
  
La sigla *VADER* sta per __*Valence Aware Dictionary e Sentiment Reasoner*__.

VADER è un modello pre-addestrato di Sentiment Analisys basato su regole, progettato per valutare la positività, la negatività e la neutralità di un testo. Questo modello è in grado di gestire caratteri speciali, abbreviazioni ed emoji, e si basa sull'intensità del *sentiment*. Lo score finale è ottenuto dalla media dei vari score delle parole.  


L'output di VADER è una quadrupla contenente i punteggi di Positività, Negatività, Neutralità e il punteggio **Compound**, che fornisce un grado di significatività complessivo degli altri tre.    
   
![Vader](immagini/Vader.png)    

Il modello VADER è particolarmente utile per analizzare testi in cui è importante considerare il contesto e l'intonazione emotiva. La sua capacità di gestire caratteri speciali e emoji consente di catturare sfumature di sentiment al di là delle sole parole, migliorando così l'accuratezza nelle analisi di testi complessi e ricchi di contenuto emotivo.    
Essendo già pre-addestrato, è pronto per l'uso e può essere utilizzato per valutare il sentiment in diversi contesti, come social media, recensioni di prodotti o qualsiasi altro tipo di testo.  
 
*Vader usa 5 euristiche per fare sentiment analysis :* 
- Punteggiatura
- Capitalizzazione
- Modificatori di grado
- Cambio di polarità
- Negazioni        


**1. Punteggiatura :** il modello da significato diverso alle stesse frasi se hanno una diversa punteggiatura.   
Se si ha un sentimento positivo si sommano i coefficienti al sentiment della parola:    
- **0.292 per ogni punto esclamativo**      
- **0.18 per ogni punto interrogativo**      
Se si riscontra un sentiment negativo si sottraggono gli stessi valori .     

**2. Capitalizzazione :** il modello tiene conto di ogni parola scritta totalmente in maiuscolo.    
Se si ha un sentimento positivo si sommano i coefficienti al sentiment della parola:       
- **0.733 per ogni punto esclamativo**  
Se si riscontra un sentiment negativo si sottrae la stessa quantità.        

**3. Modificatori di grado :** VADER ha un dizionario di booster e smorzatori del grado del sentiment.  
La presenza di un modificatore :    
- **aggiunge 0.293 al sentiment score della frase se positivo**     
sottrae la stessa quantità se il sentiment è negativo.       

**4. Cambio di polarità con il “ma” :** si ha un checker che tiene conto dei “ma”.  
Ogni volta che si incontra un “ma” si ha un cambio di polarità :     
- **il sentiment di tutte le parole prima del “ma” è decrementato del 50%**     
- **il sentiment di tutte le parole dopo del “ma” è incrementato del 50%**          

**5. Negazioni :** per ogni parola controlla il trigramma precedente a ciascuna parola.     
Se trova una negazione nelle tre parole precedenti :    
- **decrementa lo score del sentiment di quella parola di 0.74**

### Pulizia dei testi
Poiché il wrapper sporca un po' i testi delle canzoni, per lavorare in modo ordinato estraggo i testi ed formatto il tutto utilizzando le Regex, e successivamente salvo il tutto in un nuovo file JSON.

```python
import json
import re

def clean_text(text):
    # Divido il testo in righe
    lines = text.split('\n')
    
    # Rimuovo la prima riga che mi dice il numero di contributors
    cleaned_lines = lines[1:]
    
    # Controllo se l'ultima riga contiene "embed" e un numero che la precede
    if cleaned_lines[-1].lower().endswith('embed'):
        cleaned_lines[-1] = re.sub(r'\d*embed$', '', cleaned_lines[-1], flags=re.IGNORECASE).strip()
    
    
    cleaned_text = '\n'.join(cleaned_lines)
    
    # eliminazione stringhe tra [ ]
    cleaned_text = re.sub(r'\[.*?\]', '', cleaned_text)
    
    return cleaned_text.strip()


#file contenente le canzoni formattate
clean_lyrics = "clean_lyrics.json"
data = {
    "songs" : []
}

for i in range(len(artist.songs)):
    cleaned_text = clean_text(artist.songs[i].lyrics)
    data["songs"].append({
        "title": artist.songs[i].title,
        "lyrics": cleaned_text
    })

# Scrivo sul json
with open(clean_lyrics, 'w') as json_file:
    json.dump(data, json_file, indent=4)
```

### Calcolo frequenza di una parola
Lazza è solito dire la parola "ehi" in ogni sua canzone, calcoliamo la frequenza della parola all'interno dei suoi testi.

```python
import json
import re

with open('clean_lyrics.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Funzione per ricerca della parola
def count_word_frequency(text, word='ehi'):
    text = text.lower()
    # Pulizia del testo
    text = clean_text(text)
    # ricerca di tutte le occorrenze della parola
    words = re.findall(r'\b{}\b'.format(re.escape(word)), text)
    return len(words)

# Pulizia del testo
def clean_text(text):
    # Rimozione dei caratteri non alfanumerici
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Sostituzione spazi doppi con singolo
    text = re.sub(r'\s+', ' ', text)
    return text

# Funzione di conteggio numero parole totali
def count_total_words(text):
    text = clean_text(text)
    # Divido il testo 
    words = text.split()
    # Conta il numero di parole
    total_words = len(words)
    return total_words

# liste di memorizzazionee
searched_word_frequencies = []
total_word_counts = []

# Ricerca parole e numero conteggio parole totali
for song in data['songs']:
    lyrics = song['lyrics']
    searched_word_frequency = count_word_frequency(lyrics)
    total_words = count_total_words(lyrics)
    
    searched_word_frequencies.append(searched_word_frequency)
    total_word_counts.append(total_words)

# Frequenza totale della parola
total_searched_word_frequency = sum(searched_word_frequencies)
# Numero totale delle parole
total_words_all_songs = sum(total_word_counts)

# Stampa dei risultati
print(f"La frequenza totale della parola è: {total_searched_word_frequency}")
print(f"Il numero totale di parole in tutte le canzoni è: {total_words_all_songs}")
percentuale = (total_searched_word_frequency / total_words_all_songs) * 100
print(f"In percentuale: {percentuale}% \n")

for song, searched_word_freq, total_words in zip(data['songs'], searched_word_frequencies, total_word_counts):
    print(f"Titolo: {song['title']}, Frequenza: {searched_word_freq} || Numero totale di parole: {total_words}")
```
